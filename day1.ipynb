Addition , subtraction and Dot product with numpy and without numpy and calculating time taken for execution

```python
import numpy as np
import pandas as pd
import time
```

```python
list1 = [5,6,8,9,12]
list2 = [11,5,7,2,14]
```

```python
start_time = time.perf_counter()
```

```python
def add_vectors(l1,l2):
  if len(l1) != len(l2):
    raise ValueError("Vectors must have the same length")
  return [x + y for x,y in zip(l1,l2)]
```

```python
result = add_vectors(list1,list2)
result
```

```python
def sub_vectors(l1,l2):
  if len(l1) != len(l2):
    raise ValueError("Vectors must have the same length")
  return [x - y for x,y in zip(l1,l2)]
```

```python
result = sub_vectors(list1,list2)
result
```

```python
# Dot product
dot_product = sum(x * y for x,y in zip(list1,list2))
```

```python
dot_product
```

```python
end_time = time.perf_counter()
```

```python
execution_time = end_time - start_time
```

```python
print(f"Time taken: {execution_time:.8f} seconds")
```

with numpy

```python
vector1 = np.array([5,6,8,9,12])
vector2 = np.array([11,5,7,2,14])
```

```python
start_time_numpy = time.perf_counter()
```

```python
vector_add = np.add(vector1,vector2)
vector_add
```

```python
vector_subtract = np.subtract(vector1,vector2)
vector_subtract
```

```python
dot_product = np.dot(vector1 , vector2)
dot_product
```

```python
end_time_numpy = time.perf_counter()
```

```python
execution_time_numpy = end_time_numpy - start_time_numpy
```

```python
print(f"Time taken for numpy:{execution_time_numpy:.8f} seconds")
```

Numpy is faster than Python

```python
speed = execution_time / execution_time_numpy
print(f"Numpy is {speed:.2f} times faster than Python")
```
