```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('student_scores.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
numeric_df = df.select_dtypes(include=np.number)
```

```python
numeric_df.head()
```

```python
print("\nNumeric Data:")
print(numeric_df)
```

```python
# Convert to NumPy array
matrix = numeric_df.to_numpy()
```

```python
print("\nNumPy Matrix:")
print(matrix)
```

```python
matrix.shape
```

```python
# Manual Transpose (without using .T)
rows, cols = matrix.shape

manual_transpose = []

for col in range(cols):
    new_row = []

    for row in range(rows):
        new_row.append(matrix[row][col])

    manual_transpose.append(new_row)

manual_transpose = np.array(manual_transpose)

print("\nManual Transpose:")
print(manual_transpose)
```

```python
# NumPy Transpose for verification
numpy_transpose = matrix.T
```

```python
print(numpy_transpose)
```
