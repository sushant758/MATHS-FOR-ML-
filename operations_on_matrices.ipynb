# operations on Matrices using Numpy

```python
import numpy as np
import time
```

```python
# test and compare performance of numpy vs plain ppython loop - vector addition.

n=10000000
a=[1]*n
b=[2]*n

start = time.time()

c=[0]*n


for i in range(n):
  c[i] = a[i] + b[i]

end = time.time()

print(end-start)
```

```python
# now performing the same task using numpy
start = time.time()

a_np = np.arange(n)
b_np = np.arange(n)

c_np = a_np + b_np  #

end = time.time()
print(end-start)
```

```python
# create 2 metrices using numpy array
A = np.array([[1,2,3],[0,1,0],[0,0,1]]) # matrix A
B = np.array([[0,1,0],[1,0,0],[1,2,1]]) # matrix B

# vectorization is being used internally here,bcz it is converted into
# numpy array.
```

```python
# find sum of the metrices
print(A+B)
```

```python
# find differnce of the metrices
print(A-B)
```

```python
# product
print(A@B)
```

```python
print(np.matmul(A,B))
```

```python
# transpose
print(A.T)
```

```python
print(np.transpose(A))
```

```python
# determinant
print(np.linalg.det(A))
```

```python
print(np.linalg.det(B))
```

```python
# inverse
print(np.linalg.inv(A))
```

```python
print(np.linalg.inv(B))
```

```python
# Trace Find

print(np.linalg.trace(A))
```

```python
print(np.linalg.trace(B))
```

```python
# rank of a matrix
print(np.linalg.matrix_rank(A))
```

```python
print(np.linalg.matrix_rank(B))
```

```python
# eigens value and eigen vector of a matrix

print(np.linalg.eig(A))
```
