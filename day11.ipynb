```python
import numpy as np

v = np.array([4, 3])

onto = np.array([5, 0])
```

```python
dot_vu = np.dot(v, onto)

print(dot_vu)
```

```python
onto_mag_sq = np.dot(onto, onto)

print(onto_mag_sq)
```

```python
projection_scalar = dot_vu / onto_mag_sq

print(projection_scalar)
```

```python
projection = projection_scalar * onto

print(projection)
```

```python
def project(v, onto):

    dot_vu = np.dot(v, onto)

    onto_mag_sq = np.dot(onto, onto)

    projection_scalar = dot_vu / onto_mag_sq

    projection = projection_scalar * onto

    return projection
```

```python
proj = project(v, onto)

print(proj)
```

```python
import matplotlib.pyplot as plt

proj = project(v, onto)

plt.figure(figsize=(6,6))

plt.quiver(
    0,0,
    v[0],v[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='v'
)

plt.quiver(
    0,0,
    onto[0],onto[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='onto'
)

plt.quiver(
    0,0,
    proj[0],proj[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='projection'
)

plt.xlim(0,6)
plt.ylim(0,5)

plt.grid()
plt.legend()

plt.show()
```
