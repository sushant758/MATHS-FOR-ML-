# Linear Regression — Without sklearn


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('auto-mpg.csv')
```

```python
# =========================
# 2. Features and Target
# =========================

X = df.drop('mpg', axis=1)
y = df['mpg']

# Convert to NumPy
X = X.to_numpy()
y = y.to_numpy()
```

```python
# =========================
# 3. Train-Test Split
# =========================

n_samples = len(X)

indices = np.arange(n_samples)

np.random.seed(42)

np.random.shuffle(indices)

train_size = int(0.8 * n_samples)

train_indices = indices[:train_size]
test_indices = indices[train_size:]

X_train = X[train_indices]
X_test = X[test_indices]

y_train = y[train_indices]
y_test = y[test_indices]

print("Train samples:", len(X_train))
print("Test samples :", len(X_test))
```

```python
# =========================
# 4. Add Bias Column
# =========================

X_train_bias = np.c_[np.ones((X_train.shape[0], 1)), X_train]
X_test_bias = np.c_[np.ones((X_test.shape[0], 1)), X_test]
```

```python
# =========================
# 5. Normal Equation
# =========================

X_transpose = X_train_bias.T

XTX = np.dot(X_transpose, X_train_bias)

XTX_inv = np.linalg.inv(XTX)

XTy = np.dot(X_transpose, y_train)

weights = np.dot(XTX_inv, XTy)

print("\nWeights:")
print(weights)
```

```python
# =========================
# 6. Predictions
# =========================

y_pred = np.dot(X_test_bias, weights)

print("\nFirst 10 Predictions:")
print(y_pred[:10])
```

```python
# =========================
# 7. Metrics From Scratch
# =========================

# MAE
mae = np.mean(np.abs(y_test - y_pred))

# MSE
mse = np.mean((y_test - y_pred) ** 2)

# RMSE
rmse = np.sqrt(mse)

# R² Score
ss_res = np.sum((y_test - y_pred) ** 2)

ss_tot = np.sum((y_test - np.mean(y_test)) ** 2)

r2 = 1 - (ss_res / ss_tot)

print("\nEvaluation Metrics")
print("-------------------")
print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R²  :", r2)
```

```python
# =========================
# 8. Compare Actual vs Predicted
# =========================

comparison = pd.DataFrame({
    "Actual": y_test[:10],
    "Predicted": y_pred[:10]
})

print("\nActual vs Predicted")
print(comparison)
```
