# 2×2 inverse by hand find

```python
B = np.random.rand(2,2)
a = B[0][0]
b = B[0][1]
c = B[1][0]
d = B[1][1]
```

```python
B
```

```python
determinant = (a*d) - (b*c)

if determinant == 0:
  raise ValueError('Matrix is singular and cannot be inverted')

inv_matrix = [
     [ d / determinant, -b / determinant],
     [-c / determinant,  a / determinant]
]

inv_matrix
```

```python
# Verify with Numpy

det_numpy = np.linalg.det(B)

adj_matrix_numpy = np.array([
    [B[1,1], -B[0,1]],
    [-B[1,0], B[0,0]]
])

inv_matrix_numpy = adj_matrix_numpy / det_numpy

inv_matrix_numpy
```

# 3×3 inverse by hand find

```python
# Random 3x3 matrix
m = np.random.rand(3,3)

print("Original Matrix:\n")
print(m)

a = m[0,0]
b = m[0,1]
c = m[0,2]

d = m[1,0]
e = m[1,1]
f = m[1,2]

g = m[2,0]
h = m[2,1]
i = m[2,2]
```

```python
# Determinant
det = (
    a*(e*i - f*h)
    - b*(d*i - f*g)
    + c*(d*h - e*g)
)

print("\nDeterminant:")
print(det)

if np.isclose(det, 0):
    raise ValueError("Matrix is singular and cannot be inverted.")

# Cofactor Matrix
cofactor = np.array([
    [ (e*i - f*h), -(d*i - f*g),  (d*h - e*g)],
    [-(b*i - c*h),  (a*i - c*g), -(a*h - b*g)],
    [ (b*f - c*e), -(a*f - c*d),  (a*e - b*d)]
])

print("\nCofactor Matrix:")
print(cofactor)

# Adjugate = Transpose of Cofactor
adjugate = cofactor.T

print("\nAdjugate Matrix:")
print(adjugate)

# Manual Inverse
inv_matrix = adjugate / det

print("\nManual Inverse:")
print(inv_matrix)
```

```python

# NumPy Verification
numpy_inverse = np.linalg.inv(m)

print("\nNumPy Inverse:")
print(numpy_inverse)

# Check if both are equal
print("\nAre they equal?")
print(np.allclose(inv_matrix, numpy_inverse))
```

Geometric Meaning of an Inverse Matrix

The inverse matrix represents the reverse of a transformation. If a matrix stretches, rotates, or shears a vector, its inverse undoes that transformation and restores the original vector. Multiplying a matrix by its inverse results in the identity matrix, which leaves vectors unchanged.
