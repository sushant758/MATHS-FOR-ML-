# Eigenvalues & Eigenvectors


```python
import numpy as np
import pandas as pd
```

Eigenvector :

A direction that remains unchanged after a transformation.

or

An eigenvector is a special vector whose direction remains unchanged after a matrix transformation.

Eigenvalue:

The amount by which that direction is stretched or compressed.

or

An eigenvalue tells how much an eigenvector is stretched or shrunk during the transformation.

# Assignment Part 1

```python
A = np.array([
    [2, 0],
    [0, 3]
])


print(A)
```

```python
# By Hand
# Av=λv

# Where:
# A = Matrix
# v = Eigenvector
# λ = Eigenvalue


# What is happening?
# We are looking for special vectors whose direction does not change after the matrix transformation

# Note:
# Eigenvectors keep their direction after transformation;
#   eigenvalues tell how much they are stretched or compressed.
```

```python
# Step 2: Move Everything to One Side

# Mathematics = Av−λv=0
#             = (A−λI)v=0
```

```python
I = np.eye(2)

print(I)
```

```python
# Note

# We introduce the identity matrix so that λ can be written in matrix form and subtracted from A.
```

```python
print("""
A - λI =

[2-λ   0  ]
[ 0   3-λ ]
""")
```

```python
print("det(A - λI) = 0")
```

```python
print("(2 - λ)(3 - λ) = 0")
```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)
```

```python
import numpy as np

A = np.array([
    [2, 0],
    [0, 3]
])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)
```
