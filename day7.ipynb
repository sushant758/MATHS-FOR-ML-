# Determinant & Rank

```python
# For 2×2:

# A=[a b
#   c d]

# det(A)=ad−bc

# If det(A)=0 then ,
# Matrix Singular
# Matrix Inverse Does Not Exist
```

```python
# BY hand

A = np.array([
    [4, 2],
    [3, 1]
])
```

```python
det_A = ((4 * 1) - (2 * 3))
det_A
```

```python
# verify with numpy

numpy_det_A = np.linalg.det(A)
numpy_det_A
```

# Part 2: 3×3 Determinant

```python
B = np.array([
    [1, 2, 3],
    [0, 1, 4],
    [5, 6, 0]
])
```

```python
det_B = ( 1*(((1*0) -(6*4)) - 2*((0*0) - (5*4)) + 3*((0*6) - (5*1))) )
det_B
```

```python
numpy_det_B = np.linalg.det(B)
numpy_det_B
```

```python
# conclusion = The determinant tells us whether a matrix can be inverted and how much it stretches or shrinks space.
```

# Part 3: Rank

The rank indicates:

How much independent information a matrix contains.

```python
A = np.array([
    [1, 2],
    [2, 4]
])
```

```python
# here Row2 = 2 × Row1., here second row is giving a new information.  . Rank = 1
```

```python
np.linalg.matrix_rank(A)
```

### 3 by 3 matrix

```python
X = np.array([
    [1, 2],
    [2, 4],
    [3, 6],
    [4, 8]
])
```

```python
# Column2 = 2 × Column1.
```

```python
rank_X = np.linalg.matrix_rank(X)
rank_X
```

# Industry-Level Extension

```python
feature_1 = np.random.rand(100)
feature_1
```

```python
feature_2 = 2 * feature_1
```

```python
X = np.column_stack([feature_1, feature_2])
```

```python
X
```

```python
# check rank
rank_X = np.linalg.matrix_rank(X)
rank_X
```

```python
# check determinant


XTX = X.T @ X
print(XTX.shape)
np.linalg.det(XTX)
```

```python
# check inverse
pseudo_inverse = np.linalg.pinv(XTX)
pseudo_inverse
```
