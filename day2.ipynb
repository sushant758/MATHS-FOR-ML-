Write your own l1_norm() and l2_norm() functions from scratch, verify against np.linalg.norm().

```python
import numpy as np
import pandas as pd
from math import sqrt
```

```python
x = [3, 4, -2]
```

```python
# L1 Norm
# ∣3∣+∣4∣+∣−2∣ = 9
```

```python
def l1_norm(vector):
  total = 0

  for value in vector:
    total += abs(value)

  return total
```

```python
# L2 Norm
# sqrt((3*3) + (4*4) + (-2 * -2))
```

```python
sqrt(9 + 16 + 4)
```

```python
# COde:

def l2_norm(vector):
  total = 0

  for value in vector:
    total += value ** 2

  return total ** 0.5
```

```python
# Verify Against NumPy
```

```python
# L1 norm
np.linalg.norm(x, ord=1)
```

```python
# L2 Norm
np.linalg.norm(x, ord=2)
```

```python
# Comparing

x = np.array([3, 4, -2])

print("My L1:", l1_norm(x))
print("NumPy L1:", np.linalg.norm(x, ord=1))

print("My L2:", l2_norm(x))
print("NumPy L2:", np.linalg.norm(x, ord=2))
```

```python
# ==========================================
# L1 Norm (Manhattan Norm)
# ==========================================
#
# Formula:
# ||x||₁ = Σ |xᵢ|
#
# Meaning:
# 1. Take the absolute value of each element.
# 2. Add all absolute values together.
#
# Example:
# x = [3, 4, -2]
#
# ||x||₁ = |3| + |4| + |-2|
#         = 3 + 4 + 2
#         = 9
#
# ==========================================
```

```python
# ==========================================
# L2 Norm (Euclidean Norm)
# ==========================================
#
# Formula:
# ||x||₂ = √(Σ xᵢ²)
#
# Meaning:
# 1. Square each element.
# 2. Add all squared values.
# 3. Take the square root of the sum.
#
# Example:
# x = [3, 4, -2]
#
# ||x||₂ = √(3² + 4² + (-2)²)
#         = √(9 + 16 + 4)
#         = √29
#         = 5.385
#
# ==========================================
```

```python
# ============================================================
# Euclidean Distance (L2 Distance)
# ============================================================
#
# Formula:
#
#            __________________________
#           /        n
# L2 = √   /   Σ (xᵢ - yᵢ)²
#         /       i=1
#
#
# ------------------------------------------------------------
# Example
# ------------------------------------------------------------
#
# Point A = (10, 3)
# Point B = (3, 3)
#
# Step 1: Coordinate Differences
#
# x difference = 10 - 3 = 7
# y difference =  3 - 3 = 0
#
# Step 2: Square the Differences
#
# 7² = 49
# 0² = 0
#
# Step 3: Add the Squares
#
# 49 + 0 = 49
#
# Step 4: Take Square Root
#
# √49 = 7
#
# Therefore:
#
# Euclidean Distance = 7
```

From sklearn

```python
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import manhattan_distances
```

```python
X = [[0, 1],
    [1, 1]]

print("Dataset:")
print(X)

print("\nEuclidean Distance Matrix:")
print(euclidean_distances(X, X))
```

```python
origin = [[0, 0]]

print("\nDistance from Origin (0,0):")
print(euclidean_distances(X, origin))
```

```python
print("\nManhattan Distance Examples:")

print(manhattan_distances([[3]], [[3]]))
# |3 - 3| = 0

print(manhattan_distances([[3]], [[2]]))
# |3 - 2| = 1

print(manhattan_distances([[2]], [[3]]))
# |2 - 3| = 1
```

```python
A = [[1, 2],
    [3, 4]]

B = [[1, 2],
    [0, 3]]

print("\nManhattan Distance Matrix:")
print(manhattan_distances(A, B))
```
