# Matrix Multiplication

1. Matrix multiplication understanding.
2. matmul() write with nested loops.
3. Two matrices multiplication.
4. verify with np.dot().
5. Time compare

```python
import numpy as np
import time
```

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])
```

```python
print(A.shape)
```

```python
print(B.shape)
```

```python
rows_A = A.shape[0]
cols_A = A.shape[1]

print(rows_A)
print(cols_A)
```

```python
rows_B = B.shape[0]
cols_B = B.shape[1]

print(rows_B)
print(cols_B)
```

```python
if cols_A != rows_B:
    raise ValueError("Matrix multiplication not possible")
```

```python
result = np.zeros((rows_A, cols_B))
```

```python
for i in range(rows_A):
  for j in range(cols_B):
    total = 0
    for k in range(cols_A):
      A[i][k]
      B[k][j]
      total += A[i][k] * B[k][j]
      result[i][j] = total
```

# Speed Comparison

```python
A = np.random.rand(50, 50)
B = np.random.rand(50, 50)
```

```python
print(A)
```

```python
print(B)
```

```python
def matmul(A, B):

    rows_A, cols_A = A.shape
    rows_B, cols_B = B.shape

    if cols_A != rows_B:
        raise ValueError("Matrix multiplication not possible")

    result = np.zeros((rows_A, cols_B))

    for i in range(rows_A):
        for j in range(cols_B):

            total = 0

            for k in range(cols_A):
                total += A[i][k] * B[k][j]

            result[i][j] = total

    return result
```

```python
start = time.perf_counter()

my_result = matmul(A, B)

end = time.perf_counter()

print(f"Custom Time: {end - start:.6f} seconds")
```

```python
import time

matmul_start = time.perf_counter()

my_result = matmul(A, B)

matmul_end = time.perf_counter()

print("Result Shape:", my_result.shape)
print(f"Custom Time: {matmul_end - matmul_start:.6f} seconds")
```

```python
numpy_start = time.perf_counter()

numpy_result = np.dot(A, B)

numpy_end = time.perf_counter()

print("NumPy Time:", numpy_end - numpy_start)
```

```python
extra_time = (matmul_end - matmul_start) - (numpy_end - numpy_start)
print("Extra Time:", extra_time)
```
